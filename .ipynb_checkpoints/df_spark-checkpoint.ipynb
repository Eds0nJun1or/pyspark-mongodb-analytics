### Configuração do Ambiente e Leitura dos Dados

```python
# Importações Spark
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# Importações Machine Learning
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.clustering import KMeans
from pyspark.ml.classification import LogisticRegression

# Importações Análise de Dados
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import FuncFormatter
from matplotlib import font_manager

# Importações Previsão de Séries Temporais
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
# Criação da sessão Spark
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('Analise Precos Combustiveis') \
    .getOrCreate()
```

```python
# Lê o arquivo CSV
df = (
    spark
    .read 
    .option('delimiter', ';')
    .option('header', 'true')
    .option('inferSchema', 'true')
    .option('encoding', 'ISO-8859-1')
    .csv('./src/Preços semestrais - AUTOMOTIVOS_2023.02.csv')
)
```

```python
# Exibe o esquema do dataframe
df.printSchema()
```

### Tratamento e Exploração de Dados

```python
# Seleciona as colunas relevantes
df_precos = (
    df
    .select('Estado - Sigla', 'Produto', 'Valor de Compra', 'Valor de Venda', 'Unidade de Medida', 'Data da Coleta')
)
```

```python
# Verificando registros iniciais
df_precos.show(5)
```

```python
# Filtra e mostra as linhas onde o valor da coluna 'Valor de Compra' não é nulo
(
    df_precos
    .where(
        F.col('Valor de Compra').isNotNull()
    )
    .show()
)
```

```python
# Verificação de valores nulos em cada coluna
print("\nValores nulos por coluna:")
df.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```python
# Quantidade de registros e variáveis disponíveis
print(f"Quantidade de registros: {df.count()}")
print(f"Variáveis disponíveis: {len(df.columns)}")
```

```python
# Seleção e transformação de colunas
df_precos = (
    df
    .select('Estado - Sigla', 'Produto', 'Valor de Venda', 'Unidade de Medida')
    .withColumn(
        "Valor de Venda",
        F.regexp_replace(F.col("Valor de Venda"), ",", ".")
        .cast("float")
    )
)
```

```python
# Análise estatística dos preços por Estado, Produto e Unidade de Medida
df_precos_analise = (
    df_precos
    .groupBy(
        F.col('Estado - Sigla'),
        F.col('Produto'),
        F.col('Unidade de Medida')
    )
    .agg(
        F.min(F.col("Valor de Venda")).alias('menor_valor'),
        F.max(F.col("Valor de Venda")).alias('maior_valor')
    )
    .withColumn(
        "diferenca",
        F.col('maior_valor') - F.col('menor_valor')
    )
    .orderBy('diferenca', ascending=False)
)
```

```python
# Exibição dos 10 primeiros registros da análise de preços
df_precos_analise.show(10)
```

```python
# Análise de Preços por Estado
df_precos_analise = (
    df_precos
    .groupBy('Estado - Sigla', 'Produto', 'Unidade de Medida')
    .agg(
        F.min(F.col("Valor de Venda")).alias('menor_valor'),
        F.max(F.col("Valor de Venda")).alias('maior_valor'),
        F.mean(F.col("Valor de Venda")).alias('media_valor'),
        F.stddev(F.col("Valor de Venda")).alias('desvio_padrao_valor'),
        F.count(F.col("Valor de Venda")).alias('contagem')
    )
    .withColumn("diferenca", F.col('maior_valor') - F.col('menor_valor'))
    .orderBy('diferenca', ascending=False)
)
```

```python
# Exibe as primeiras linhas do dataframe resultante
df_precos_analise.show(10)
```

```python
# Identificação de Outliers (primeiro calculando a média e desvio padrão)
df_outliers_stats = (
    df_precos
    .groupBy('Produto')
    .agg(
        F.mean(F.col("Valor de Venda")).alias('media_valor'),
        F.stddev(F.col("Valor de Venda")).alias('desvio_padrao_valor')
    )
)
```

```python
# Juntar as estatísticas de volta ao dataframe original
df_precos = df_precos.join(df_outliers_stats, on="Produto", how="left")
```

```python
# Identificação de Outliers
df_outliers = (
    df_precos
    .withColumn(
        "outlier",
        F.when(
            (F.col("Valor de Venda") > F.col("media_valor") + 3 * F.col("desvio_padrao_valor")) |
            (F.col("Valor de Venda") < F.col("media_valor") - 3 * F.col("desvio_padrao_valor")),
            1
        ).otherwise(0)
    )
)
```

```python
# Exibe os outliers identificados
df_outliers.show()
```

```python
# Contagem das bandeiras em ordem crescente
df_bandeiras = (
    df
    .groupBy('Bandeira')
    .count()
    .orderBy('count', ascending=True)
)
```

```python
# Exibe as bandeiras em ordem crescente de contagem
df_bandeiras.show(23)
```

```python
# Estatísticas de preços para cada bandeira
df_bandeira_precos = (
    df
    .withColumn(
        "Valor de Venda",
        F.regexp_replace(F.col("Valor de Venda"), ",", ".").cast("float")
    )
    .groupBy('Bandeira')
    .agg(
        F.mean(F.col("Valor de Venda")).alias('preco_medio'),
        F.min(F.col("Valor de Venda")).alias('menor_preco'),
        F.max(F.col("Valor de Venda")).alias('maior_preco')
    )
    .orderBy('preco_medio')
)
```

```python
# Exibe as estatísticas de preços por bandeira
df_bandeira_precos.show(23)
```

```python
# Região com Maior Volume de Vendas
df_regiao_vendas = df.groupBy("Regiao - Sigla") \
                   .agg(F.sum("Valor de Venda").alias("Total_Vendas")) \
                   .orderBy("Total_Vendas", ascending=False) \
                   .first()

print(f"Região com maior volume de vendas: {df_regiao_vendas['Regiao - Sigla']}")
```

```python
# Ticket Médio por Região
df_regiao_ticket_medio = df.groupBy("Regiao - Sigla") \
                           .agg(F.avg("Valor de Venda").alias("Ticket_Medio")) \
                           .orderBy("Ticket_Medio", ascending=False)

df_regiao_ticket_medio.show(5)
```

```python
# Produto mais vendido
df_produto_mais_vendido = df.groupBy("Produto") \
                          .agg(F.sum("Valor de Venda").alias("Total_Vendas")) \
                          .orderBy("Total_Vendas", ascending=False) \
                          .first()

print(f"Produto mais vendido: {df_produto_mais_vendido['Produto']}")
```

```python
# Ticket Médio por Produto
df_produto_ticket_medio = df.groupBy("Produto") \
                           .agg(F.avg("Valor de Venda").alias("Ticket_Medio")) \
                           .orderBy("Ticket_Medio", ascending=False)

df_produto_ticket_medio.show(6)
```

```python
# Filtrar valores nulos na coluna 'Valor de Venda'
df_precos_filtrado = df_precos.filter(df_precos['Valor de Venda'].isNotNull())

# Análise de preços por estado e produto, removendo dados nulos de media_valor e desvio_padrao_valor
df_precos_estado_produto = (
    df_precos_filtrado
    .groupBy('Estado - Sigla', 'Produto')
    .agg(
        F.mean(F.col("Valor de Venda")).alias('media_valor'),
        F.stddev(F.col("Valor de Venda")).alias('desvio_padrao_valor')
    )
    .filter((F.col("media_valor").isNotNull()) & (F.col("desvio_padrao_valor").isNotNull()))  # Filtra valores nulos
    .orderBy('Estado - Sigla', 'Produto')
)

# Exibe as linhas do dataframe resultante
df_precos_estado_produto.show()
```

```python
# Filtra os valores nulos na coluna 'Valor de Venda' e converte para o formato adequado
df_precos_filtrado = df.filter(df['Valor de Venda'].isNotNull()) \
                       .withColumn('Valor de Venda', F.regexp_replace('Valor de Venda', ',', '.').cast('float'))
```

```python
# Exibe as primeiras linhas do DataFrame para garantir que os dados estão corretos
df_precos_filtrado.show(5)
```

```python
# Calcula a média diária dos preços de venda
df_tendencias = (
    df_precos_filtrado.groupBy('Data da Coleta')
                      .agg(F.mean(F.col("Valor de Venda")).alias('media_valor_diaria'))
                      .orderBy('Data da Coleta')
)
```

```python
# Exibe as primeiras linhas do DataFrame de tendências
df_tendencias.show()
```

```python
# Criação de uma nova view temporária 'view_precos' com transformação de dados
view_precos = spark.sql("""
    select 
        `Estado - Sigla`, 
        `Produto`, 
        regexp_replace(`Valor de Venda`, ",", ".") as `Valor de Venda`,
        `Unidade de Medida`
    from combustiveis
""")
```

```python
# Criação ou substituição da view temporária 'view_precos' e exibição dos primeiros 5 registros
view_precos.createOrReplaceTempView('view_precos')
view_precos.show(5)
```

```python
# Consulta SQL para calcular diferenças de preços e exibição dos 5 maiores valores de diferença
df_diferenca_precos = spark.sql("""
    select 
        `Estado - Sigla`,
        `Produto`,
        `Unidade de Medida`,
        MIN(`Valor de Venda`) as menor_valor,
        MAX(`Valor de Venda`) as maior_valor,
        MAX(`Valor de Venda`) - MIN(`Valor de Venda`) as diferenca
   from view_precos
   group by all
   order by diferenca DESC
""")
df_diferenca_precos.show(5)
```

```python
# Converte o DataFrame de tendências para pandas
pdf_tendencias = df_tendencias.toPandas()

# Convertendo 'Data da Coleta' para tipo datetime usando o formato correto
pdf_tendencias['Data da Coleta'] = pd.to_datetime(pdf_tendencias['Data da Coleta'], format='%d/%m/%Y')

# Renomeia as colunas para 'ds' e 'y', que são esperadas pelo Prophet
data_frame = pdf_tendencias[['Data da Coleta', 'media_valor_diaria']].rename(columns={'Data da Coleta': 'ds', 'media_valor_diaria': 'y'})

# Cria e ajusta o modelo
model = Prophet(daily_seasonality=True)
model.fit(data_frame)

# Cria um DataFrame para as datas futuras
future_dates = model.make_future_dataframe(periods=100)  # Por exemplo, para prever os próximos 30 dias

# Faz as previsões
forecast = model.predict(future_dates)

# Plota os resultados
model.plot(forecast)
plt.title('Previsão da Evolução dos Preços Médios de Venda')
plt.xlabel('Data da Coleta')
plt.ylabel('Preço Médio de Venda (R$)')
plt.grid(True)
plt.show()
```

```python
# Filtra os valores nulos na coluna 'Valor de Venda' e converte para o formato adequado
df_precos_filtrado = df.filter(df['Valor de Venda'].isNotNull()) \
                       .withColumn('Valor de Venda', F.regexp_replace('Valor de Venda', ',', '.').cast('float'))

# Calcula as tendências temporais dos preços de venda por produto
df_tendencias_produto = (
    df_precos_filtrado.groupBy('Data da Coleta', 'Produto')
                      .agg(F.mean('Valor de Venda').alias('media_valor_diaria'))
                      .orderBy('Data da Coleta')
)

# Converte para Pandas para plotagem correta
pdf_tendencias_produto = df_tendencias_produto.toPandas()

# Ajusta o tamanho da figura
plt.figure(figsize=(14, 7))

# Plotando a comparação das tendências temporais por produto
sns.lineplot(data=pdf_tendencias_produto, x='Data da Coleta', y='media_valor_diaria', hue='Produto', marker='o')
plt.title('Comparação das Tendências Temporais dos Preços de Venda por Produto')
plt.xlabel('Data da Coleta')
plt.ylabel('Preço Médio de Venda')
plt.legend(title='Produto', loc='upper left')
plt.grid(True)

# Exibe o gráfico
plt.show()
```

```python
# Calcula as tendências temporais dos preços de venda por estado
df_tendencias_estado = (
    df_precos_filtrado.groupBy('Data da Coleta', 'Estado - Sigla')
                      .agg(F.mean('Valor de Venda').alias('media_valor_diaria'))
                      .orderBy('Data da Coleta')
)

# Converte para Pandas para plotagem correta
pdf_tendencias_estado = df_tendencias_estado.toPandas()

# Obtém a lista de estados únicos
estados = pdf_tendencias_estado['Estado - Sigla'].unique()

# Ajusta o tamanho da figura
plt.figure(figsize=(16, 20))

# Plotando a comparação das tendências temporais por estado em gráficos separados
for i, estado in enumerate(estados, start=1):
    plt.subplot(len(estados)//2 + len(estados)%2, 2, i)
    data_estado = pdf_tendencias_estado[pdf_tendencias_estado['Estado - Sigla'] == estado]
    sns.lineplot(data=data_estado, x='Data da Coleta', y='media_valor_diaria', marker='o')
    plt.title(f'Tendência Temporal dos Preços de Venda - {estado}')
    plt.xlabel('Data da Coleta')
    plt.ylabel('Preço Médio de Venda')
    plt.grid(True)

# Ajusta o layout da figura para evitar sobreposição
plt.tight_layout()

# Exibe os gráficos
plt.show()
```
